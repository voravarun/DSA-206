# Read JSON

## Convert CSV to JSON

```python
def csv_to_json(csvFilePath, jsonFilePath):
    jsonArray = []
      
    #read csv file
    with open(csvFilePath, encoding='utf-8') as csvf: 
        #load csv file data using csv library's dictionary reader
        csvReader = csv.DictReader(csvf) 

        #convert each csv row into python dict
        for row in csvReader: 
            #add this python dict to json array
            jsonArray.append(row)
  
    #convert python jsonArray to JSON String and write to file
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf: 
        jsonString = json.dumps(jsonArray, indent=4)
        jsonf.write(jsonString)
          
csvFilePath = r'fishdata1.csv'
jsonFilePath = r'fishdata1.json'
csv_to_json(csvFilePath, jsonFilePath)
```

# Creation of arrays

```python
import pandas as pd
df1 = pd.DataFrame({
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']},
    index=[0,1,2,3])
```

```python
df1
```

```python
df2 = pd.DataFrame({
    'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']},
    index=[4,5,6,7])
```

```python
df2
```

```python
df3 = pd.DataFrame({
    'A':['A8','A9','A10','A11'],
    'B':['B8','B9','B10','B11'],
    'C':['C8','C9','C10','C11'],
    'D':['D8','D9','D10','D11']},
    index=[8,9,10,11])
```

```python
df3
```

# Concatenation

```python
frames = [df1,df2,df3]
frames
```

```python
result = pd.concat(frames)
result
```

# Horizontal Merge

```python
df4 = pd.DataFrame({
    'B':['B2','B3','B6','B7'],
    'D':['D2','D3','D6','D7'],
    'F':['F2','F3','F6','F7']},
    index=[2,3,6,7])
df4
```

```python
result = pd.concat([df1,df4], axis=1,sort=False)
result
```

# Inner join

```python
result = pd.concat([df1,df4], axis=1,join='inner')
result
```

# Outer join

```python
result = pd.concat([df1,df4], axis=1,join='outer')
result
```

# Append

```python
result = df1.append(df2)
result
```

## Append multiple dataframe

```python
result = df1.append([df2,df3])
result
```
